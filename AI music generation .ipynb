# Pre-processing all the files 

```python
notes = []
for file in glob.glob("midi_files/*.mid"):
    midi = converter.parse(file)
    elements_to_parse = midi.flat.notes
    print("parsing %s"%(file))
    
    for ele in elements_to_parse:
        # if the element is a note then we will store it's pitch
        if isinstance(ele, note.Note):
            notes.append(str(ele.pitch))


        # if the ele is a chord we will append it's normal order 
        elif isinstance(ele, chord.Chord):
            notes.append('+'.join(str(n) for n in ele.normalOrder))
    
    
```

```python
with open('notes','wb') as filepath:
    pickle.dump(notes, filepath)
```

```python
with open('notes', 'rb') as f:
    notes =pickle.load(f)
```

```python
print(len(set(notes)))
vocab_len= len(set(notes))
```

```python
print("total number of unique notes %d"%(len(set(notes))))
print("total number of notes in total %d"%(len(notes)))
```

# Prepare sequential data for LSTM

```python
# How many elements should you consider as input
sequence_len= 100
```

```python
pitch_name = sorted(set(notes))
```

```python
#mapping the element to the corresponding integer
ele_to_int = dict((ele, num) for num , ele in enumerate(sorted(set(notes))))
ele_to_int
```

```python
network_input = []
network_output = []

```

```python
 for i in range( len(notes) - sequence_len):
        seq_in = notes[i:i+sequence_len]
        seq_out = notes[i+sequence_len]
        
        
        network_input.append([ele_to_int[ch] for ch in seq_in])
        network_output.append(ele_to_int[seq_out])
```

```python
total_patterns = len(network_input)
network_input = np.reshape(network_input, ( total_patterns, sequence_len , 1 ) ) 
```

```python
normalised_network_input = network_input/ float(vocab_len)
print( normalised_network_input.shape)
```

```python
network_output = to_categorical(network_output) 
print(network_output.shape)
print(type(network_output))
```
